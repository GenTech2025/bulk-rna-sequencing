```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the raw_counts.txt file
raw_counts = pd.read_csv("../../02_data/processed-data/raw_counts.txt", delimiter="\t", comment="#")
```

```python
# Check the dimensions of the data frame
raw_counts.shape
```

```python
raw_counts.head(5)
```

The dimension of the data frame is 78899 rows and 33 columns, our total number of samples were 27, hence we can remove the 6 remaining columns.

```python
# Set the rowname of the data frame to gene ids
raw_counts = raw_counts.set_index("Geneid")

# Create a new data frame and only keep the columns corresponding to the different samples in our dataset
counts = raw_counts.iloc[:,5:] # df.iloc[rows, columns]

# Check the new data frame
counts.head(5)
```

After we have gotten rid of the unecessary columns, now we focus on renaming the columns so that each column is named as the run accession for the sample.

```python
# Use python regular expression to rename the column to their suitable name.
counts.columns = counts.columns.str.extract(r"(SRR\d+)", expand=False)

# Remove the index name
counts.index.name = None

# Check the new column names
counts.columns
```

```python
# Double check the number of samples
len(counts.columns)
```

```python
# Write out the counts data frame as a csv file

# Source - https://stackoverflow.com/a
# Posted by Andy Hayden, modified by community. See post 'Timeline' for change history
# Retrieved 2025-12-15, License - CC BY-SA 4.0
counts.to_csv("../../02_data/processed-data/counts.csv", sep='\t', encoding='utf-8', index=True, header=True)
```
