#### How to actually use the ***entrez-direct*** tool to link the run accessions to their BioSamples records?

```bash
# Create a folder to store intermediate files
mkdir -p ../../02_data/intermediate-files

# Use the command below to check the different fields available
esearch -db sra -query PRJNA603669 \
| efetch -format runinfo \
| head -n 1 \
> ../../02_data/intermediate-files/available_fields.txt
```
